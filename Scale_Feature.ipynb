```python
import pandas as pd
```

```python
df = pd.read_csv('/content/cars2.csv')
```

```python
df
```

```python
x = df[['Weight','Volume']]
```

```python
y = df[['CO2']]
```

```python
from sklearn import linear_model
```

```python
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
```

```python
dx = scale.fit_transform(x)
```

```python
regr = linear_model.LinearRegression()
```

```python
regr.fit(dx,y)
```

```python
scale = scale.transform([[2300,1.3]])
```

```python
predictedCO2 = regr.predict([scale[0]])
```

```python
print(predictedCO2)
```
